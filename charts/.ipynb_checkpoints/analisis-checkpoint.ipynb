```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readFrom(contenido, beg):
    # Iterar a través del contenido carácter por carácter
    save=0;
    for i in range(beg, len(contenido)):
        # Verificar si el carácter actual es un dígito o un punto decimal
        if contenido[i].isdigit() or contenido[i] == '.':
            # Si es así, construir una cadena que representa el número
            num_str = contenido[i]
            for j in range(i + 1, len(contenido)):
                if contenido[j].isdigit() or contenido[j] == '.':
                    num_str += contenido[j]
                else:
                    save = j 
                    break
            # Convertir la cadena a un valor de punto flotante y salir del bucle
            num = float(num_str)
            #print("El próximo número flotante es:", num)
            return [num, save];
            
    
    
```

# Evaluacion de Polinomios

```python
cantSizes = 15; # de 10000 a 150000
eval1 = np.zeros(cantSizes)
eval2 = np.zeros(cantSizes)
eval3 = np.zeros(cantSizes)
eval4 = np.zeros(cantSizes)
```

```python
f = open('polinomios.txt','r')
contenido = f.read()

idx = 0
for j in range(cantSizes):
    eval1[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    eval2[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    eval3[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    eval4[j], idx = readFrom(contenido,idx)
```

```python
print(eval2)
```

```python
sizes = np.zeros(cantSizes, int)
i=1;
for i in range(1,cantSizes+1):
    sizes[i-1] = i*10000
```

```python
plt.plot(sizes, eval1, 'o-', label='Evaluacion 1')
plt.plot(sizes, eval2, 's-', label='Evaluacion 2')
plt.plot(sizes, eval3, 's-', label='Evaluacion 3')
plt.plot(sizes, eval4, 's-', label='Evaluacion 4')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Evaluacion de polinomios Execution Time')
plt.legend()
plt.show()
```

# Suma iterativa y recursiva

```python
cantSizes = 32; # de 1000 a 32000
suma = np.zeros(cantSizes)
rsum = np.zeros(cantSizes)

f = open('suma.txt','r')
contenido = f.read()

idx = 0
for j in range(cantSizes):
    suma[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    rsum[j], idx = readFrom(contenido,idx)
    
sizes = np.zeros(cantSizes, int)
i=1;
for i in range(1,cantSizes+1):
    sizes[i-1] = i*1000
```

```python
sizes
```

```python
plt.plot(sizes, suma, 's-', label='Suma')
plt.plot(sizes, rsum, 's-', label='Suma recursiva')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Suma Execution Time')
plt.legend()
plt.show()
```

# Search

```python
cantSizes = 10; # de 100000000 a 190000000
seq = np.zeros(cantSizes)
bin1 = np.zeros(cantSizes)
bin2 = np.zeros(cantSizes)

f = open('search.txt','r')
contenido = f.read()

idx = 0
for j in range(cantSizes):
    seq[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    bin1[j], idx = readFrom(contenido,idx)

for j in range(cantSizes):
    bin2[j], idx = readFrom(contenido,idx)
    
sizes = np.zeros(cantSizes, int)
i=1;
for i in range(1,cantSizes+1):
    sizes[i-1] = i*100000000
```

```python
plt.plot(sizes, seq, 'o-', label='Sequential')
plt.plot(sizes, bin1, 's-', label='binary 1')
plt.plot(sizes, bin2, 's-', label='binary 2')
plt.xlabel('Sizes')
plt.ylabel('Time')
plt.title('Search Execution Time')
plt.legend()
plt.show()
```
